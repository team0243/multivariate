ให้เขียน function คำนวณจำนวนวินาทีจาก เวลาต่อไปนี้

'14-15-31'

'13-00-01'

'05-49-44'

'00-00-16'

และ print ให้สวยงาม

````
def time(t):
  j3=  int (t.split('-')[0])*3600
  k3= int (t.split('-')[1])*60
  n3 = int (t.split('-')[2])
  tt = j3+k3+n3
  print("แปลงเป็นวินาทีได้เท่ากับ",tt," วินาที")
````

````
time('14-15-31') 
````

````
time('13-00-01') 
````

````
time('05-49-44')
````

````
time('00-00-16')  
````

***Inverse matrix***

ถ้า A เป็นเมตริกซ์ที่มีมิติ n x n จะพบว่า

  AIn = InA = A

เราจึงเรียก In ว่าเป็นเอกลักษณ์การคูณของเมตริกซ์ และถ้าสามารถหาเมตริกซ์ B ที่ทำให้

 AB = BA = In

 เราจะเรียก B ว่าเป็น อินเวอร์สการคูณของ A
